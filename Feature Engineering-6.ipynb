### Q1. Pearson Correlation Coefficient Calculation

The **Pearson correlation coefficient** (r) is calculated using the formula:

\[
r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n \sum x^2 - (\sum x)^2][n \sum y^2 - (\sum y)^2]}}
\]

**Example Data**:
| Study Time (Hours) | Exam Score |
|---------------------|------------|
| 2                   | 70         |
| 3                   | 75         |
| 4                   | 80         |
| 5                   | 85         |
| 6                   | 90         |

**Calculations**:
- \(n = 5\)
- \(\sum x = 20\) (Total study time)
- \(\sum y = 400\) (Total exam scores)
- \(\sum xy = 2 \times 70 + 3 \times 75 + 4 \times 80 + 5 \times 85 + 6 \times 90 = 1450\)
- \(\sum x^2 = 2^2 + 3^2 + 4^2 + 5^2 + 6^2 = 70\)
- \(\sum y^2 = 70^2 + 75^2 + 80^2 + 85^2 + 90^2 = 28825\)

Substituting the values into the formula:

\[
r = \frac{5(1450) - (20)(400)}{\sqrt{[5(70) - (20)^2][5(28825) - (400)^2]}} = \frac{7250 - 8000}{\sqrt{[350 - 400][144125 - 160000]}} = \frac{-750}{\sqrt{[-50][-15875]}}
\]

Calculating this gives \( r \approx 0.997 \).

**Interpretation**:
A Pearson correlation coefficient of approximately 0.997 indicates a very strong positive linear relationship between study time and exam scores.


### Q2. Spearman's Rank Correlation Calculation

The **Spearman's rank correlation coefficient** (ρ) is calculated using the formula:

\[
ρ = 1 - \frac{6 \sum d^2}{n(n^2 - 1)}
\]

where \(d\) is the difference between the ranks of each variable.

**Example Data**:
| Sleep (Hours) | Job Satisfaction |
|----------------|------------------|
| 6              | 8                |
| 5              | 7                |
| 8              | 9                |
| 4              | 6                |
| 7              | 8                |

**Ranked Data**:
| Sleep (Rank) | Job Satisfaction (Rank) | d (Difference) | d² |
|---------------|-------------------------|----------------|----|
| 4             | 3                       | 1              | 1  |
| 5             | 2                       | 1              | 1  |
| 6             | 5                       | -1             | 1  |
| 7             | 4                       | 3              | 9  |
| 8             | 1                       | 4              | 16 |

Calculating \(d^2\):
- \(\sum d^2 = 28\)
- \(n = 5\)

Substituting into the formula:

\[
ρ = 1 - \frac{6 \times 28}{5(5^2 - 1)} = 1 - \frac{168}{120} = 1 - 1.4 = -0.4
\]

**Interpretation**:
A Spearman's rank correlation of -0.4 indicates a moderate negative monotonic relationship between sleep and job satisfaction, suggesting that as sleep decreases, job satisfaction tends to decrease as well.
